```python
import numpy as np
# Из списка
mass = np.array([1, 2, 3, 4, 5])
print("Массив из списка:", mass)
zeros_arr = np.zeros(3)  # массив из нулей
one_arr = np.ones((2, 2))  # матрица 2x2 из единиц
range_arr = np.arange(0, 6)  # последовательность 0-5
print("Zeros:", zeros_arr)
print("Ones:\n", one_arr)
print("Arange:", range_arr)

print("Shape:", mass.shape)  # форма массива
print("Dtype:", mass.dtype)  # тип данных

print("mass[2] =", mass[2])  # третий элемент
print("Срез mass[1:4] =", mass[1:4])  # элементы с 1 по 3

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print("a + b =", a + b)  # поэлементное сложение
print("a * 2 =", a * 2)  # умножение на число

data = np.array([1, 2, 3, 4, 5])
print("Mean:", np.mean(data))  # среднее значение
print("Sum:", np.sum(data))    # сумма
```
